```python
# 不同时间的订单数量
time_trans=pd.DataFrame(data,columns=['order_date','Unnamed: 0']).groupby(by='order_date').count()
time_trans.columns=['trans_count']

time_trans_7=time_trans.rolling(7).mean()
plt.figure(figsize=(15,6))
plt.plot(time_trans_7.index,time_trans_7.trans_count)
plt.title('Number of transactions for each month', fontsize=14)
x_major_locator=MultipleLocator(30)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
```

```python
# 不同时间的销售总额
time_amount=pd.DataFrame(data,columns=['order_date','mean_amount']).groupby(by='order_date').sum()
time_amount_7=time_amount.rolling(7).mean()
plt.figure(figsize=(15,6))
plt.plot(time_amount_7.index,time_amount_7.mean_amount)
plt.title('Total amount for each month', fontsize=14)
x_major_locator=MultipleLocator(30)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
time_amount
```
