```python
#### import the package here

import requests, json
import pandas as pd
import ndjson
import time 
import numpy as np 
import matplotlib.pyplot as plt

#### Finding out the optimal batch size for the loop

time_count =[]

url = "https://opendatabank-apigw.hktvmall.com/gw/api/v1/PublicSaleTransactionSimplified"

for _ in [10**i for i in range(4,8)]:
#     print(_)
    start_time = time.time()

    batch_size = int(_)

    payload={
        "sys_take": f"{batch_size}",
        "sys_skip": "0"
    }
    headers = {
      'Authorization': f'Bearer {token}',
      'Content-Type': 'application/json'
    }

    response = requests.request("POST", url, headers=headers, json=payload)
    
    data_list = response.json(cls=ndjson.Decoder)
    
    pd.DataFrame(data_list)
    
    time_count.append((time.time() - start_time)/_)
    
```
