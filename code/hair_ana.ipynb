```python
import numpy as np
import pandas as pd
import datetime
# from datetime import timedelta
from matplotlib.pyplot import MultipleLocator
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/Users/heart_jf/Desktop/5005/finalB/hair.csv')
data
```

```python
# data['order_date']=[datetime.strptime(data.order_date[i],"%Y-%m-%d") for i in range(len(data))]
data['date_fomat']=[datetime.datetime.strptime(data.order_date[i],"%Y-%m-%d") for i in range(len(data))]
data.sort_values(by=['date_fomat'],inplace=True)
data.index=[i for i in range(len(data))]

amount=data.total_price.values.tolist()
min_amount=[float(amount[i].split('-')[0]) for i in range(len(amount))]
max_amount=[float(amount[i].split('-')[1]) for i in range(len(amount))]
mean_amount=[(min_amount[i]+max_amount[i])/2 for i in range(len(amount))]
data['mean_amount']=mean_amount

data.head()
```

```python
# 不同时间的订单数量
time_trans=pd.DataFrame(data,columns=['date_fomat','order_number']).groupby(by='date_fomat').count()
plt.figure(figsize=(15,6))
plt.plot(time_trans.index,time_trans.order_number)
plt.title('Number of transactions for each month', fontsize=14)
x_major_locator=MultipleLocator(30)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)

# time_trans
```

```python
# 不同时间的销售总额
time_amount=pd.DataFrame(data,columns=['date_fomat','mean_amount']).groupby(by='date_fomat').sum()
plt.figure(figsize=(15,6))
plt.plot(time_amount.index,time_amount.mean_amount)
plt.title('Total amount for each month', fontsize=14)
x_major_locator=MultipleLocator(30)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
# time_amount
```

```python
# 不同地点的销售订单
loc_trans=pd.DataFrame(data,columns=['district','order_number']).groupby(by='district').count()
loc_trans
```

```python
# 不同地点的销售总额
loc_amount=pd.DataFrame(data,columns=['district','mean_amount']).groupby(by='district').sum()
loc_amount.columns=['total_amount']
loc_amount.head(50)
loc_amount
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="test")
```

```python
loc=loc_amount.index.values.tolist()
location=[]
count=0
for i in loc:
    if count%10==0:
        print(count)
    count+=1
    location.append(geolocator.geocode(i))
```

```python
loc_trans['location']=location
loc_trans.dropna(inplace=True)
loc_trans
```

```python
loc_amount['location']=location
loc_amount.dropna(inplace=True)
loc_amount
```

```python
latitude=[]
longitude=[]
location=loc_amount.location
for i in range(len(location)):
    latitude.append(location[i].latitude)
    longitude.append(location[i].longitude)
```

```python
loc_trans['latitude']=latitude
loc_trans['longitude']=longitude
loc_amount['latitude']=latitude
loc_amount['longitude']=longitude
```

```python
loc_trans.to_csv('location_transaction.csv')
```
